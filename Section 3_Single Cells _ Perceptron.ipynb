# ***Perceptron***

# INTRODUCTION:



1. Perceptron is a single layer feed forward neural network designed by Rosenblatt and Minsky Papert (in 1962).
2. Perceptron is a linear classifier (binary). Also, it is used in supervised learning. It helps to classify the given input data.
3. It consists of sensory unit(input), associator unit(hidden), response unit(output).
4. Sensory/input units have weights assigned as a  binary activation function (-1,0,1) assigned at random . 
5. The binary step with fixed 'θ'(theta) is used as activation for the associator unit. The output signals sent from the associator unit to response unit are always binary.
6. The output of a perceptron is given by - 

            y = f(x)

            where activation function is - 

            f(x) = 1 ,  if x > θ

                 = 0 ,  if -θ <= x <= θ

                 = -1 ,  if x < -θ

# WORKING:



1. The above described activation function learning is used in weight updation between associator unit and response unit.
2. For each training input, the net will calculate the response and it will determine whether or not an error has occured.
3. Error calculation is based on the comparison of value of targets with calculated outputs.
4. Weights on the connections are adjusted suitably - 

        w(new) = w(old) + xαt

        b(new) = b(old) + αt 

        where   t -> target value (1 or -1)

                α -> learning rate

                b -> bias

perceptron-6168423.jpg

# ALGORITHM:



STEP 0 : Initialize the weights and the bias. Also initialize the learning rate 'α' such that 0 < α <=1.

STEP 1 : Perform STEP 2-6 until the final stopping  condition  is false.

STEP 2 : Perform STEP 3-5 for each training pair indicated by s : t

STEP 3 : The input layer containing input units is applied with identity activation  functions:  x = s

STEP 4 : Calculate the output of the network. To do so, find the net input: 

        y = b + Σxw ( for i = 1 to n)
        
        where n -> number of neurons

then applying activation over net input:

      y = f(x)
 
     where activation function is - 
 
     f(x) = 1 ,  if x > θ
 
          = 0 ,  if -θ <= x <= θ
 
          = -1 ,  if x < -θ


STEP 5 :  Perform weight and bias adjustment. Compare the value of actual (calculated) output and desired (target) output.

         w(new) = w(old) + xαt
 
         b(new) = b(old) + αt 

STEP 6 : Train the network until there is no weight change. This is the stopping condition for the network. If condition is not true , then start from STEP 2.


      


# ACTIVATION FUNCTION:

1. In Artificial Neural Networks (ANN) , the activation function of a node defines the output of that node given an input or set of inputs.
2. The activation function decide whether a neuron should be activated or not. 
3. The activation function is a non-linear transformation that we do over the input before sending it to the next layer of neurons or finalizing it as output.
4. There can be several different types of activation like  - 

  a. Step-activation-function  (easiest one)

  b. Sigmoid-activation-function (widely used)

  c. ReLU-activation-function (widely used)

  d. Leaky ReLU-activation-function (improved version of ReLU)
  

Activationfunction.jpg

1. The above image represents the Step activation function and Sigmoid activation function.
2. Lets say we are performing a binary classification the output may either be 0 or 1 forr which step function can be easily applied. In this, we consider a threshold value and if the value of net input is greater than the threshold then the neuron is activated.
3. Say a misclassification took place and an error e has occured while classifying then bias and weights have to be updated as mentioned  in the algorithm above. Though a minor updation in weights or bias may result into an entirely different output as the inputs are binary.
4. In order to rectify such misclassifications we have sigmoid activation function and as shown in the image the classification is gradual i.e. regression. Rectifying misclassification becomes easier by tweaking weights and bias without any drastic change in output.

1. The sigmoid activation function is a smooth function and is continuously differentiable. 
2. The biggest advantage that it has over step and linear function is that it is non-linear. 
3. This is an incredibly cool feature of the sigmoid function. This essentially means that when I have multiple neurons having sigmoid function as their activation function – the output is non linear as well. The function ranges from 0-1 having an S shape.

# CREATING PERCEPTRON MODEL

```
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.linear_model import Perceptron
```

```
iris = load_iris()
```

```
type(iris)
```

```
iris
```

We will create a perceptron model which would identify whether the flower is setosa or not using petal width and petal length (independent variables). 

```
x = iris.data[:, (2,3)]
```

```
x
```

```
iris.target
```

```
y = (iris.target == 0)
```

```
y
```

```
y = (iris.target == 0).astype(np.int)
```

```
y
```

```
x
```

```
classifier = Perceptron(random_state=42)  # hyperparameter not necessary
```

```
classifier.fit(x,y)
```

```
y_pred = classifier.predict(x)
```

```
y_pred
```

```
classifier.coef_ # line coefficient
```

```
classifier.intercept_ # line intercept
```

```
from sklearn.metrics import accuracy_score
```

```
accuracy_score(y, y_pred)
```
